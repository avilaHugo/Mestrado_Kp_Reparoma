```python
# Define the power law function
def power_law(x, k, gamma):
    return k * np.power(x, gamma)

def pangenome_openess(gamma):
    print(gamma)
    if 0 < gamma < 1:
        return 'Aberto'
    if gamma < 0:
        return 'Fechado'

    raise ValueError

rare_species = species_rarefaction_data.specie.unique()
n_species = rare_species.shape[0]
n_iterations = species_rarefaction_data.it.max()
fit_by='median'

# # init a fig obje to hold all the plots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# remove unused axes
for ax in axes.flat[n_species:]:
    fig.delaxes(ax)

for i, (specie, gdf) in enumerate(species_rarefaction_data.groupby('specie')):
    # get the ax to plot
    ax = axes.flat[i]

    # add boxplot to the ax
    boxplot_data = (
        gdf
        .drop(['it', 'specie'], axis=1)
        .sort_values(['n_genomes', 'pan'])
    )


    line_data = (
        boxplot_data
        .groupby('n_genomes')
        .agg([fit_by])
    )

    popt, pcov = curve_fit(power_law, line_data.index, line_data.pan[fit_by])

    
    boxplot_data.plot(x='n_genomes', y='pan', kind='scatter', ax=ax, marker='o', color='#8fc1ff', s=3)
    
    ax.scatter(line_data.index, line_data.pan[fit_by], color='red', marker='o', s=4)
    
    ax.plot(line_data.index, power_law(line_data.index, *popt), label='power law', color='#313233')
    
    # ax.set_ylim(4000, 13000)
    

    ax.set_title(f'{specie}')
    

    # Add Gamma value to the plot
    ax.text(
        0.05, 0.95, f'Œ≥: {popt[1]:.3f} - Pangenoma: { pangenome_openess(popt[1]).lower() }', transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5)
    )

    # remove x and y labels
    ax.set_xlabel('')
    ax.set_ylabel('')


# add title to the figure
_ = fig.suptitle('Curvas de rarefa√ß√£o das esp√©cies do g√™nero Klebsiella', fontsize=16)

# add a legend to the figure 
_ = fig.legend(
    labels=[
        'Permuta√ß√µes (10 itera√ß√µes)',
        'Medianas',
        'Ajuste de curva (M√©todo dos M√≠nimos Quadrados)'
    ],
    loc='lower left',
    bbox_to_anchor=(0.4, 0.25),
    bbox_transform=fig.transFigure,
    ncol=1,
)

# add a title to the x axis with fontsize 14
_ = fig.text(0.5, 0.04, 'N√∫mero de genomas amostrados por permuta√ß√£o', ha='center', fontsize=14)
_ = fig.text(0.04, 0.5, 'N√∫mero de clusters', va='center', rotation='vertical', fontsize=14)
```

```python
def to_partition(x):
    assert 0 <= x <= 1, 'x must be between 0 and 1'
    if x >= 0.85:
        return 'Persistente' 

    if x >= 0.15 < 0.85:
        return 'Intermedi√°rio (Shell)'
    
    if x < 0.15:
        return 'Raro (Cloud)'




pan_freq = pd.concat(
    pangenome_clusts_pres_abs
    | select(
        lambda x: (
            pangenome_clusts_pres_abs[x]
            .pipe(lambda df_: df_.sum(axis=1) / df_.shape[1])
            .rename('freq')
            .to_frame()
            .assign(
                specie=x,
                partition = lambda df_: df_.freq.apply(to_partition).astype('category')
            )
            .reset_index()
        )
    ),
    ignore_index=True
)

pan_freq.head()
```

```python
# Pres/abs of genus clusters in each species

genus_presence_absence = (
    pd.concat(
        Path(pangenomes_dir).iterdir()
        | where(lambda x:  x.is_dir())
        | select(lambda x: x.glob('matrix.csv'))
        | chain
        | where(lambda x: 'pasteurii' in x.parents[0].name)
        | select(lambda x: (
                pd.read_csv(x, dtype='str')
                .loc[
                    :,
                    lambda df_: ~df_.columns.isin(df_.iloc[:, :14])
                ]
                .agg(lambda x: set(x.dropna().apply(lambda x: x.replace('gnl|Prokka|', '').replace('"', '').split()).explode()))
                .to_frame()
                .reset_index()
                .rename(columns={0: 'locus_tag', 'index': 'genome_id'})
                .explode('locus_tag')
                .set_index('locus_tag')
                .assign(
                    genus_clusters = lambda df_: df_.index.to_series().map(species_members_to_genus_clusters),
                    species = x.parents[0].name.replace('Klebsiella_', 'K. ')
                )
                .reset_index()
            )
        )

    )
    .pivot_table(
        index=['species', 'genome_id'],
        columns='genus_clusters',
        values='locus_tag',
        aggfunc='count',
    )
    .notna()
)


genus_presence_absence.to_pickle('/home/hugo.avila/Projects/reparoma/results/GWAS/features/genus_presence_absence.pkl')
genus_presence_absence
```

```python
species_genome_count = genus_presence_absence.index.get_level_values('species').value_counts().to_dict()

# get frequency of all clusters in each species
genus_presence_absence_freq = (
    genus_presence_absence
    .groupby('species')
    .apply(lambda df_: df_.sum() / species_genome_count[df_.name])
    .T
)

genus_presence_absence_freq
```

```python
reparoma = (
    pd.read_csv(
        '/home/hugo.avila/Projects/reparoma/results/GOIs/50_70/reparoma/reparoma.tsv',
        sep='\t',
        header=None,
        names=['cluster', 'target', 'pident', 'fident', 'qcov', 'tcov', 'alnlen']
    )
    .loc[
        lambda df_: (
                df_.pident.ge(0.5) & df_[['qcov', 'tcov']].ge(0.7).all(axis=1) & df_.alnlen.ge(100)
            )
    ]
    .groupby('cluster', as_index=False)
    .apply(pd.DataFrame.head, n=1)
    .droplevel(1)
    .merge(
        pd.read_csv(
            '/home/hugo.avila/Projects/reparoma/results/GOIs/50_70/reparoma/50_70.kofam.unique.parsable.txt',
            sep=';',
            header=None,
            index_col=0,
            usecols=[0, 1],
            names=['cluster', 'KO']
        ),
        left_on='cluster',
        right_on='cluster',
        how='left',
        validate='1:1',
    )
    .fillna('N/A')
    .loc[:, ['cluster', 'target', 'KO']]
    .assign(target = lambda df_: (x := df_.target) + x.str.contains(':').map({True: ' - k', False: ' - u'}))
    .merge(
        genus_presence_absence_freq,
        left_on='cluster',
        right_index=True,
        validate='1:1'
    )
    .set_index(['cluster', 'target', 'KO'])
    .loc[
        lambda df_: (
            df_.loc[:, df_.columns.str.contains('K. ')].ge(0.01).sum(axis=1).ge(2) &
            df_.index.to_frame().KO.isin(kegg_reparoma.index.to_frame().KO)
        )
    ]
    .assign(
        gene = lambda df_: df_.index.to_frame().KO.map(kegg_names).fillna('N/A'),
    )
    .set_index('gene', append=True)
)

display(reparoma)

g_reparoma = sns.clustermap(
    (df_ := reparoma),
    cmap=palette,
    yticklabels=False,
    figsize=(3, 6),
    vmin=min(df_.values.tolist() | chain | where(lambda x: x)),
)

# Set the title
g_reparoma.figure.suptitle('Reparoma', y=1.05)

# set labels
g_reparoma.ax_heatmap.set_xlabel('Esp√©cies')
g_reparoma.ax_heatmap.set_ylabel('clusters de genes do reparoma')

# add a legend "frequ√™ncia" to the right of the plot
g_reparoma.cax.set_position([1.05, .4, .05, .3])
g_reparoma.cax.set_ylabel('Frequ√™ncia')
```

```python
from collections import defaultdict

def add_number_copy(series_):
    gene_names = {}
    result = []
    for gene in series_:
        if gene in gene_names:
            gene_names[gene] += 1
        else:
            gene_names[gene] = 1

        result.append(f'{gene} ({gene_names[gene]})')
    return result
    

temp = (
    reparoma.reset_index()
    .assign(
        gene = lambda df_: df_[['gene', 'cluster' ]].apply(lambda x: f'{x[0]} ({x[1]})', axis=1),
    )
    .drop(['cluster', 'KO', 'target'], axis=1)
    .set_index('gene')
)


g_test = sns.clustermap(
    temp,
    cmap=palette,
    yticklabels=True,
    figsize=(6, 15),
    vmin=min(temp.values.tolist() | chain | where(lambda x: x)),
    dendrogram_ratio=(.2, .025),
)

# change font size yticklabels
_ = g_test.ax_heatmap.set_yticklabels(g_test.ax_heatmap.get_yticklabels(), fontsize=8)


# Set the title
_ = g_test.figure.suptitle('Reparoma', y=1.05)

# set labels
_ = g_test.ax_heatmap.set_xlabel('Esp√©cies')
_ = g_test.ax_heatmap.set_ylabel('clusters de genes do reparoma')

# add a legend "frequ√™ncia" to the right of the plot
_ = g_test.cax.set_position([1.05, .5, .03, .1])
_ = g_test.cax.set_ylabel('Frequ√™ncia')
```
